손실 함수(Loss Function)

- 타깃 값과 예측 값을 뺀 다음 제곱한 것으로 수식은 다음과 같음, SE = (y-ŷ)²

1) 가중치(w)에 대해 제곱 오차 미분

편미분 : 미분의 대상 변수(w)를 제외한 다른 변수(x, b 등)를 상수로 취급하여 미분한 것

image.png

- y는 준비된 타깃 데이터이므로 w의 함수가 아니고, ŷ은 w의 함수(ŷ=w*x+b)인 점만 주의

- 따라서 y는 그대로 두고 -ŷ에 대해서만 미분, ŷ을 w에 대해 미분하면 상수항 b는 사라지고 x만 남음. 결과는 다음과 같음


image.png

- 보통 제곱 오차 공식을 2로 나눈 함수를 편미분 함 (이유는 2와 1/2가 서로 약분되어 조금 더 깔끔하게 (-(y-ŷ)x)로 표현되기 때문)

- 손실함수에 상수를 곱하거나 나누어도 최종 모델의 가중치나 절편에 영향을 주지 않음

image.png

- 오차 역전파를 하면서 적용했던 수식(w + w_rate * err)은 제곱 오차를 미분한 것과 동일

2) 절편에 대하여 제곱 오차 미분하기

image.png

- ŷ = w * x + b 이므로, ŷ을 b에 대해 미분하면 w * x 항은 사라지고 b의 계수인 1만 남음

- 가중치에서 변화율을 뺏던 이유와 동일하게 절편에서 변화율을 뺌

image.png

※ 손실 함수에 대해 매번 변화율의 값을 계산하는 대신, 편미분을 사용하여 변화율을 계산
※ 변화율은 그레이디언트(Gradient, 경사)라고 표기
